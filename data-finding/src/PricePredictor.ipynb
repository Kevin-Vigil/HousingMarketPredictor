```python
import matplotlib.pyplot as plot
import numpy as np
import csv
import datetime
import pandas as pd

arrXAxis =[]
arrYAxis =[]

#Same CSV as shown in SRC, just relative path on local machine.
df = pd.read_csv("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/77433data.csv")
df['date'] = pd.to_datetime(df["date"], yearfirst = True)

#Graph a CSV given an X and Y axis index
def graphXY(filePath, xIndex: int, yIndex: int):
    df.plot(kind = "scatter", x = "date", y = "median_price")
    plot.show()
graphXY("77433data.csv", 0, 4)
```

```python
#Preprocessing
cols = ["postal_code", "location_name", "active_listing_count", "median_days_on_market","median_sqft","average_price","total_listing_count","quality_flag"]
df = df.drop(cols, axis=1)
df.info()

#Either drop or create normalized values for null values
#Dropping because median listing price is important
df = df.dropna()
df.info()
```

```python
from sklearn.model_selection import train_test_split

#transfer to usable data


df.date = pd.to_datetime(df.date)
x = df.date.values
y = df['median_price'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

plot.plot(x_train, y_train, 'ro')
plot.title("Train data")
plot.show()

plot.plot(x_test, y_test, 'ro')
plot.title("Test data")
plot.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()# Create the model
x_train = x_train.reshape(-1,1) # Reshape the data to a 2D array as it is currently a 1D array
y_train = y_train.reshape(-1,1)
reg.fit(x_train, y_train)#Fit the model and use training data

y_pred = reg.predict(df.date.values.astype(float).reshape(-1,1)) #Predict off the entire dataset and add prediction to the dataframe
df['pred'] = y_pred

ax = df.plot(x='date', y='median_price', color='black', style='.')#Plot the data
df.plot(x='date', y='pred', color='orange', linewidth=3, ax=ax, alpha = 0.5)
ax.set_title("Prediction of Zip code 77433")
ax.set_xlabel("Date")
ax.set_ylabel("Median Price")
reg.score(df.date.values.astype(float).reshape(-1,1),df['median_price'].values)
```

While this is not a good example of linear regression, the model may work for other zip codes. Let's convert our model into a function and run it against all other data on my local machine.

```python
import matplotlib.pyplot as plot
import numpy as np
import csv
import datetime
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

def linearReg(file_path: str, drop_cols: list, zip_name: str) -> None:
    """
    
    :param file_path: relative path to the data file
    :param drop_cols: columns to drop from the dataset
    :param zip_name: name of the zip code that Linear Regression is being trained on.
    """
    df = pd.read_csv(file_path)
    df['date'] = pd.to_datetime(df["date"], yearfirst = True)

    #Preprocessing
    df = df.drop(drop_cols, axis=1)
    df = df.dropna()

    df.date = pd.to_datetime(df.date)
    x = df.date.values
    y = df['median_price'].values
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

    reg = LinearRegression()# Create the model
    x_train = x_train.reshape(-1,1) # Reshape the data to a 2D array as it is currently a 1D array
    y_train = y_train.reshape(-1,1)
    reg.fit(x_train, y_train)#Fit the model and use training data
    y_pred = reg.predict(df.date.values.astype(float).reshape(-1,1)) #Predict off the entire dataset and add prediction to the dataframe
    df['pred'] = y_pred
    ax = df.plot(x='date', y='median_price', color='black', style='.')#Plot the data
    df.plot(x='date', y='pred', color='orange', linewidth=3, ax=ax, alpha = 0.5)
    ax.set_title("Prediction of Zip code " + zip_name)
    ax.set_xlabel("Date")
    ax.set_ylabel("Median Price")
    print("prediction score of zip code " + zip_name + ":", reg.score(df.date.values.astype(float).reshape(-1,1),df['median_price'].values))

cols = ["postal_code", "location_name", "active_listing_count", "median_days_on_market",
        "median_sqft","average_price","total_listing_count","quality_flag"]
    
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/01027data.csv", cols, "01027")
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/49802data.csv", cols, "49802")
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/61378data.csv", cols, "61378")
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/76471data.csv", cols, "76471")
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/83204data.csv", cols, "83204")
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/93422data.csv", cols, "93422")
linearReg("../data/RDC_Inventory_Core_Metrics_Zip_History_processed_files/22472data.csv", cols, "22472")
```

As you can see with some scores and graphs, some zip codes do not fare very well with Linear Regression, while some may be an amazing fit to the model for predicting future housing market prices.
